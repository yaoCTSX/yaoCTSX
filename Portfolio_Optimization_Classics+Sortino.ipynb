## Returns

```python
def returns(df):
    return np.log(df/df.shift(1))
```

```python
returns_df = returns(data1)
```

```python
returns_avg = returns_df.mean()*100
```

## Covarience Matrix

```python
def covarience_matrix(returns_df):
    return returns_df.cov()
```

```python
cov_mat = covarience_matrix(returns_df)
```

```python
cov_mat.head()
```

## Expected ratio and Standard Deviation

```python
e_r = data1.resample('Y').last().pct_change().mean()
e_r
```

```python
sd = data1.pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(len(data1)))
sd
```

```python
assets = pd.concat([e_r, sd], axis=1)
assets.columns = ['Returns', 'Volatility']
assets
```

## Simulation of portfolios

```python
p_ret = []
p_vol = []
p_weights = []


num_assets = len(data1.columns)
num_portfolios = 50000
```

```python
weights = np.random.random(num_assets)
weights/np.sum(weights)
```

```python
%%time
for portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)
    returns = np.dot(weights, e_r)
    p_ret.append(returns)
    var = cov_mat.mul(weights, axis=0).mul(weights, axis=1).sum().sum()
    sd = np.sqrt(var)
    ann_sd = sd*np.sqrt(len(data1))
    p_vol.append(ann_sd)
```

```python
data = {'Returns':p_ret, 'Volatility':p_vol}
```

```python
for counter, symbol in enumerate(data1.columns.to_list()):
    data[str(symbol)] = [w[counter] for w in p_weights]
```

```python
portfolios = pd.DataFrame(data)
portfolios
```

```python
portfolios.plot.scatter(x='Volatility', y='Returns', grid=True)
```

## Sortino Ratio

```python
def sortino_ratio(series, N,rf):
    mean = series.mean() * N -rf
    std_neg = series[series<0].std()*np.sqrt(N)
    return mean/std_neg
```
